```python
import pandas as pd
```

```python
complete_data = pd.read_csv('../data/curated/complete_data.csv')
complete_data
```

```python
data = complete_data[['Postcode','Baths','Population', 'Count_School', 'Count_Station','Bedrooms','Weekly Rental Price']]
data['Count_School'] = data['Count_School'].fillna(0)
data['Count_Station'] = data['Count_Station'].fillna(0)
data = data.dropna(axis=0)

data
```

```python
print(data['Weekly Rental Price'].min())
print(data['Weekly Rental Price'].max())
```

```python
import matplotlib.pyplot as plt

plt.boxplot(data['Weekly Rental Price'])
plt.xlabel('Weekly Rental Price')

plt.show()
```

```python
rent = data['Weekly Rental Price'].value_counts()

pd.DataFrame(data = {'min': [rent.min()],
                     'max': [rent.max()],
                     'mean': [rent.mean()],
                     'std': [rent.std()]})
```

```python
Q0 = data['Weekly Rental Price'].quantile(0.05)
Q1 = data['Weekly Rental Price'].quantile(0.25)
Q3 = data['Weekly Rental Price'].quantile(0.75)
Q4 = data['Weekly Rental Price'].quantile(0.95)
IQR = Q3 - Q1
print("Q0: ", Q0)
print("Q1: ", Q1)
print("Q3: ", Q3)
print("Q4: ", Q4)
print("IQR: ", IQR)
```

```python
# select the data that has rent from 125 to 805 per week

data = data[((data['Weekly Rental Price'] >= (Q1 - 1.5 * IQR)) & (data['Weekly Rental Price'] <= (Q3 + 1.5 * IQR)))]

data
```

```python
import matplotlib.pyplot as plt

plt.boxplot(data['Weekly Rental Price'])
plt.xlabel('Weekly Rental Price')

plt.show()
```

```python
data = data[(data['Baths'] > 0) & (data['Bedrooms'] > 0)]

data
```

```python
data.to_csv('../data/curated/outliers_removed.csv')
```
