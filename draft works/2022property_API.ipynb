```python
import requests
import json
import pandas as pd

client_id = 'client_b7c03c3f8e9a5711d92e47a1c4d30b1f'
client_secret = 'secret_949906594bbbac023858403c1fa7f4fe'
scopes = 'api_listings_read'
auth_url = 'https://auth.domain.com.au/v1/connect/token'
url_endpoint = 'https://api.domain.com.au/v1/listings/residential/_search'

def get_property_info(client_id, client_secret, scopes, auth_url, url_endpoint):
    '''The functrion takes client_id, client_secret, scopes, auth_url and url_endpoint, in order to request data from domain,
    it would convert json data to dataframe, and present data.'''
    
    #Performs a POST on the specified url to get the access token
    response = requests.post(auth_url, data = {
                        'client_id':client_id,
                        'client_secret':client_secret,
                        'grant_type':'client_credentials',
                        'scope':scopes,
                        'Content-Type':'text/json'
                        })
    #Extracting data in json format
    json_res = response.json()
    #parse the json to get the access token
    access_token=json_res['access_token']
    auth = {'Authorization':'Bearer ' + access_token}
    
    # sending post request and saving the response as response obj
    res1 = requests.post(url_endpoint, headers=auth, json={
        "listingType":"Rent",
        "pageSize":200,
        "pageNumber":1,
        "minBedrooms":1,
        "minBathrooms":1,
        "minCarspaces":0,
        "locations":[
            {
            "state":"VIC",
            "region":"",
            "area":"",
            "suburb":"",
            "postCode":"",
            "includeSurroundingSuburbs":""
            }
        ]
    }) 
    #Extracting data in json format
    r = json.loads(res1.text)
    #Convert the data into dataframe
    df = pd.json_normalize(r)
    return df

#call the function
get_property_info(client_id, client_secret, scopes, auth_url, url_endpoint)
```
