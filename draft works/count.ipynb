```python
import pandas as pd

# Read in bus station information
station = pd.read_csv('../data/curated/bus_station_postcode.csv')
station = station.drop(columns=['Unnamed: 0'])
station
```

```python
count_station = station.groupby('Postcode').count()

count_station = count_station.drop(columns=['Bus_Station_Name'])

count_station = count_station.rename(columns={'Location' : 'Count_Station'})

count_station
```

```python
# read the hospital

hospital = pd.read_csv('../data/raw/hospital.csv')

hospital = hospital[['Formal Name', 'Postcode']]

hospital_count = hospital.groupby('Postcode').count()

hospital_count = hospital_count.rename(columns={'Formal Name' : 'Count_Hospital'})

hospital_count
```

```python
# read the school

school = pd.read_csv('../data/raw/schoollocations2022.csv')

school = school[['School_Name', 'Postal_Postcode']]

school = school.rename(columns={'Postal_Postcode' : 'Postcode'})

school_count = school.groupby('Postcode').count()

school_count = school_count.rename(columns={'School_Name' : 'Count_School'})

school_count
```

```python
# read the shopping center

shopping_center = pd.read_csv('../data/curated/shopping_center_postcode.csv')

shopping_center = shopping_center[['Shopping_Center_Name', 'Postcode']]

shopping_center_count = shopping_center.groupby('Postcode').count()

shopping_center_count = shopping_center_count.rename(columns={'Shopping_Center_Name' : 'Count_Shopping_Center'})

shopping_center_count
```

```python
count = pd.merge(count_station, hospital_count, how='left', on='Postcode')

count = pd.merge(count, school_count, how='left', on='Postcode')

count = pd.merge(count, shopping_center_count, how='left', on='Postcode')

count = count.fillna(0)

count.to_csv('../data/curated/count.csv')
```
