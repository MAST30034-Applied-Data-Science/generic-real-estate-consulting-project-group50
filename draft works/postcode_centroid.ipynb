```python
import pandas as pd

centroid = pd.read_csv('../data/raw/australian_postcodes(1).csv')
centroid = centroid[['postcode','locality','state','long','lat']]
centroid = centroid[((centroid['postcode'] >= 3000) & (centroid['postcode'] <= 3999)) | ((centroid['postcode'] >= 8000) & (centroid['postcode'] <= 8999))]
centroid = centroid[centroid.state.isin(['VIC'])]

centroid
```

```python
from pandas.core.frame import DataFrame

c_list = []

for i in range(0,len(centroid)):
    row = centroid.iloc[i]
    location = str(row[4]) + ' ' + str(row[3])
    c_list.append([row[0],location])

centroid_new = DataFrame(data = c_list, columns=['Postcode', 'Centroid'])

centroid_new.drop_duplicates(subset=['Postcode'], keep='first', inplace=True)

centroid_new

```

```python
centroid_new.to_csv('../data/curated/centroid.csv')
```
