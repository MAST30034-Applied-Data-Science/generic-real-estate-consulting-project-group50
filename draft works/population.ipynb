```python
import pandas as pd

postcode_SA2 = pd.read_csv('../data/raw/postcode_SA2.csv')

postcode_SA2 = postcode_SA2[['POSTCODE', 'SA2_MAINCODE_2011', 'SA2_NAME_2011']]

postcode_SA2 = postcode_SA2.rename(columns={'POSTCODE' : 'Postcode', 'SA2_MAINCODE_2011' : 'SA2_code', 'SA2_NAME_2011' : 'SA2_name'})

postcode_SA2 = postcode_SA2.iloc[1763:3207]

postcode_SA2
```

```python
population = pd.read_csv('../data/raw/population.csv')

population = population[population['S/T name'] == 'Victoria']

population = population.drop(columns=['S/T code', 'S/T name', 'GCCSA code', 'GCCSA name', 'SA4 code', 'SA4 name', 'SA3 code', 'SA3 name'])

population = population.rename(columns={'SA2 code' : 'SA2_code', 'SA2 name' : 'SA2_name'})

population
```

```python
population_postcode = pd.merge(postcode_SA2, population, how='left', on=['SA2_code', 'SA2_name'])

population_postcode = population_postcode.dropna()

population_postcode
```

```python
population_postcode = population_postcode.drop(columns=['SA2_code', 'SA2_name', 'Change 2011-2021', 'Change %', 'Area km2', 'Population density 2021 persons/km2'])

population_postcode = population_postcode.groupby(['Postcode']).sum()

population_postcode.to_csv('../data/raw/population_postcode.csv')
```
