```python
#download historical data by URL
import pandas as pd
workbook_url = 'https://www.dhhs.vic.gov.au/moving-annual-rents-suburb-march-quarter-2021-excel'
def combine_func(workbook_url):
    all_df = pd.read_excel(workbook_url, sheet_name=None)
    df = pd.concat(all_df, ignore_index=True)
    return df
combine_func(workbook_url)
```
