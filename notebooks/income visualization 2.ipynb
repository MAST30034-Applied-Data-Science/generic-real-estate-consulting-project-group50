```python
#import package and libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#Reading the dataset
df = pd.read_csv('predict_income.csv')
df.head()
```

```python
data_2016 = df['2016'].median()
data_2017 = df['2017'].median()
data_2018 = df['2018'].median()
data_2019 = df['2019'].median()

data_2020 = 0
data_2021 = 0
data_2022 = 0
data_2023 = 0
data_2024 = 0
data_2025 = 0
```

```python
name_list = ["2016","2017","2018","2019","2020","2021","2022","2023","2024","2025"]
```

```python
value = [data_2016,data_2017,data_2018,data_2019,data_2020,data_2021,data_2022,data_2023,data_2024,data_2025]
```

```python
#Plotting barplot
plt.figure(figsize=(10, 5))
style.use('ggplot')
plt.title('Income Data')
plt.bar(name_list, value,color=['pink','pink','pink','pink','pink','pink','pink','pink','pink','pink'],width=0.5)
plt.xticks(rotation=50)
plt.tick_params(axis='y',which='major',labelsize=14)
plt.show()
```
